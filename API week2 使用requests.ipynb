```python
import requests
import json
KEY = 'e0faf9db606d4c19965f0416498dadce'
BASE_URL = 'https://westcentralus.api.cognitive.microsoft.com/face/v1.0/detect'
HEADERS = {
    'Content-Type':'application/json',
    'Ocp-Apim-Subscription-Key': '{}'.format(KEY),
}
img_url = 'https://timgsa.baidu.com/timg?image&quality=80&size=b9999_10000&sec=1585822877177&di=27da847d1e127015ca66ae73387c550b&imgtype=0&src=http%3A%2F%2F05imgmini.eastday.com%2Fmobile%2F20180906%2F20180906131803_63ffc80556a920bb7c2ec47cfc5f1ffc_2.jpeg'
data = {
    'url': '{}'.format(img_url),
}
payload = {
    'returnFaceId': 'true',
    'returnFaceLandmarks': 'false',
    'returnFaceAttributes': '{}'.format('age,gender,glasses,emotion'),
}
r = requests.post(BASE_URL, data=json.dumps(data), params=payload, headers=HEADERS)
r.status_code
results = r.json()
results
```
